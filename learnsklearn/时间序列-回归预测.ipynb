```python
import pandas as pd
df=pd.read_csv('~/Downloads/customerservice.csv',sep=',')
df.head(5)
```

```python
from sklearn import linear_model as lm
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
train_data = df.iloc[:,:3].values
train_target = df.iloc[:,-1].values
for i in range(30):
    X_train, X_test, y_train, y_test = train_test_split(train_data,train_target,test_size=0.2)
    lr = lm.LinearRegression()
    lr.fit(X_train,y_train)
    print(lr.score(X_test,y_test),lr.coef_,lr.intercept_)
```
