```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
X_mos, y_mos = make_moons(n_samples=100,random_state=123)#半月形数据
from sklearn.datasets import make_circles
X_cis, y_cis = make_circles(n_samples=1000,
                   random_state=123, noise=0.1,factor=0.2)
def pltshow(X,y):
    plt.scatter(X[y==0,0],X[y==0,1],
               color='red',marker='^',alpha=0.5)
    plt.scatter(X[y==1,0],X[y==1,1],
               color='blue',marker='o',alpha=0.5)
    plt.show()
    
pltshow(X_mos,y_mos)
```

```python
pltshow(X_cis,y_cis)
```

通过核PCA转化非线性可分数据至线性可分，使其能够被线性模型处理

```python
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components=2,
                kernel='rbf',gamma=15)
kpca.fit(X_cis)
X_cisnpca = kpca.transform(X_cis)
pltshow(X_cisnpca,y_cis)
lr = LogisticRegression(solver='lbfgs',multi_class='auto')
lr = lr.fit(X_cisnpca,y_cis)
print (lr.score(X_cisnpca,y_cis))
```
