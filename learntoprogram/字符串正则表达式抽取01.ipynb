```python
import re
```

```python
with open('/Users/fannian/Documents/document_config/data/票房预估一期场次时间解析.txt','r') as f:
    i = 1
    print(type(f))
    while i>0:   
        s =f.readline()
        i = len(s)
        prog = re.compile('([^0-9]*)(20[0-9][0-9])[-年.]([0-9]{1,2})[-月.]([0-9]{1,2})')
        result = prog.match(s)
        print('-'*20)
        print(s)
        if result is not None:
            print(result.group(0)+'\t'+result.group(2)+'\t'+result.group(3)+'\t'+result.group(4))
```
